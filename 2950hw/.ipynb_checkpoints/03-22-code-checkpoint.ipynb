## Single linear regression

```python
# trips as a function of temperature

temperature_model = LinearRegression()
temperature_model.fit(citibike[ ['avg_temp'] ], citibike['trips'])
print(temperature_model.coef_, temperature_model.intercept_)
```

```python
# plot fit
plt.scatter(citibike.avg_temp, citibike.trips, alpha=0.8)
plt.plot(citibike.avg_temp, temperature_model.predict(citibike[['avg_temp']]), c='darkorange')
plt.tight_layout()
plt.show()
```

## Score our simple model

Recall the **coefficient of determination**, $r^2$:

$$r^2 = (corr)^2 = \left( \frac{Cov(X,Y)}{\sqrt{Var(X)}\sqrt{Var(Y)}} \right)^2$$

We can interpret $r^2$ as **the fraction of the variation in the output variable that can be explained (or predicted) by variation in the input variable(s)**.

### Using NumPy

```python
# numpy wants a transposed input matrix (swap rows and columns)
np.cov(citibike[['trips', 'avg_temp']].T, ddof=0) # ddof=0 means N-weighted (not N-1)
```

```python
citibike[['avg_temp', 'trips']].cov(ddof=0).iloc[0,1]
```

```python
citibike['avg_temp'].var(ddof=0)
```

```python
# score the model by 'hand'

covariance_matrix = np.cov(citibike[['avg_temp', 'trips']].T, ddof=0)
print("Covariance matrix")
print(covariance_matrix)

cov_xy = covariance_matrix[0,1]
var_x  = covariance_matrix[0,0]
var_y  = covariance_matrix[1,1]

r2 = (cov_xy / np.sqrt(var_x) / np.sqrt(var_y)) **2
print(f'\nr^2 by hand: {r2}')
```

```python
# square the correlation coefficient
r = np.corrcoef(citibike[['avg_temp', 'trips']].T)[0,1]
print(f'Squared correlation coefficient: {r**2}')
```

```python
# sklearn .score() method
score = temperature_model.score(citibike[['avg_temp']], citibike['trips'])
print(f'scikit-learn r^2 score: {score}')
```

### Using Pandas

We could do all this in Pandas, too. Note that Pandas doesn't require the matrix transposition for the covaraince input and that it sometimes differs from NumPy in selecting default `ddof`.

A single example:

```python
# covariance matrix in pandas
pandas_cov_mat = citibike[['avg_temp', 'trips']].cov(ddof=0)
display(pandas_cov_mat)
```

## Multiple linear regression

In single linear regression, the output variable is modeled as a linear function of a single input variable:

$$f(x) = mx + b + \varepsilon$$

But we aren't limited to single input variables. In general, a linear regression models an output variable as a linear combination of $n$ input variables:

$$f(x) = m_1x + m_2x + \ldots + m_nx + b + \varepsilon$$

In the case of our Citibike data, what factors beyond temperature might help us predict ridership?

```python
# regress trips on avg_temp and precip
multi_model = LinearRegression()
multi_model.fit(citibike[['avg_temp', 'precip']], citibike['trips'])
print('trips | temp, precip:', multi_model.coef_, multi_model.intercept_)
```

```python
# recall the temperature-only values
print('trips | temp:', temperature_model.coef_, temperature_model.intercept_)
```

```python
# plot the new (better?) model
plt.scatter(citibike.avg_temp, citibike.trips, alpha=0.8)
plt.scatter(
    citibike.avg_temp,
    multi_model.predict(citibike[['avg_temp', 'precip']]), 
    c='darkorange',
    alpha=0.8
)
plt.tight_layout()
plt.show()
```

```python
# score the new model
multi_model.score(citibike[['avg_temp', 'precip']], citibike['trips'])
```

## Working with binary inputs

Is ridership impacted by weekends?

Create a new column in our dataframe with a 0 for weekdays and a 1 for weekends ...

```python
# the easy way, given that we already have day names
citibike['weekend'] = citibike['weekday'].apply(lambda x: 1 if x in ['Saturday', 'Sunday'] else 0)
citibike.head(7)
```

```python
# or use the datetime index
# user check: why does the lambda function work here?
citibike['weekend'] = citibike.index.weekday.map(lambda x: x//5)
```

```python
citibike.head(7)
```

```python
# plot trips | weekend
plt.figure(figsize=(6,4))
plt.scatter(
    citibike.weekend, 
    citibike.trips, 
    alpha=0.5
)
plt.tight_layout()
plt.show()
```

```python
# better, a swarmplot (or other jittered plot)
plt.figure(figsize=(6,4))
sns.swarmplot(
    x='weekend',
    y='trips',
    data=citibike
)
plt.show()
```

```python
# best(?), a boxplot
plt.figure(figsize=(6,4))
sns.boxplot(
    x='weekend',
    y='trips',
    data=citibike
)
plt.show()
```

```python
# fit a model of trips | temp, weekend
weekend_model = LinearRegression().fit(citibike[['avg_temp', 'weekend']], citibike['trips'])
print('Weekend model parameters:', weekend_model.coef_, weekend_model.intercept_)
```

```python
# score the weekend model
weekend_model.score(citibike[['avg_temp', 'weekend']], citibike['trips'])
```

## Colinear inputs

Note that our modeled coefficient for `avg_temp` changed a little when we also modeled `precip` or `weekend`. That's because some (small) portion of the variance in `avg_temp` was also captured by the other input variable. Not very much, because weekendness doesn't really change the temperature (or vice versa), and precipitation is only very weekly coupled to temperature (in New York City).

Let's check this by calculating the correlation coefficients of these input variables:

```python
citibike[['avg_temp', 'precip', 'weekend']].corr()
```

But what about variables that are more strongly correlated? Recall our correlation heatmap:

```python
# visualize correlations
sns.heatmap(citibike.corr(), center=0, cmap='coolwarm')
plt.tight_layout()
plt.show()
```

`avg_temp` and `max_temp` are *very* strongly correlated:

```python
citibike[['avg_temp', 'max_temp']].corr()
```

What happens if we use these two variables as our inputs to predict ridership? Guesses about the $r^2$ score? About the coefficients?

```python
# fit a model with correlated inputs
correlated_model = LinearRegression().fit(citibike[['avg_temp', 'max_temp']], citibike['trips'])
print('Correlated model parameters:', correlated_model.coef_, correlated_model.intercept_)
```

```python
# score the correlated model
correlated_model.score(citibike[['avg_temp', 'max_temp']], citibike['trips'])
```

Why did the score increase by so little? Because there's very little additional information in the `max_temp` data that's not already present in the `avg_temp` data.

Why did the coefficient of `avg_temp` change so much? Because the model is indifferent to *where* it gets its predictive information. Notice that the *sum* of the the coefficients in this case is very similar to the original coefficient of `avg_temp`:

```python
sum(correlated_model.coef_)
```

## One last model

Let's use all of our (non-redundant) features to build our best linear model of ridership.

We'll use `avg_temp`, `precip`, and binarized inputs for `weekend` and `snow`.

```python
# Add binary variable for snow
citibike['any_snow'] = citibike.snow.apply(lambda x: 1 if x >= 0.01 else 0)
citibike.head()
```

```python
# fit and score
best_model = LinearRegression()
best_model.fit(citibike[['avg_temp', 'precip', 'weekend', 'any_snow']], citibike['trips'])
r2_best = best_model.score(citibike[['avg_temp', 'precip', 'weekend', 'any_snow']], citibike['trips'])
print("r^2:", r2_best)
print('coefficients:', best_model.coef_)
print('intercept:', best_model.intercept_)
```

```python
# plot the new (better?) model
plt.scatter(citibike.avg_temp, citibike.trips, alpha=0.8)
plt.scatter(
    citibike.avg_temp,
    best_model.predict(citibike[['avg_temp', 'precip', 'weekend', 'any_snow']]), 
    c='darkorange',
    alpha=0.5
)
plt.tight_layout()
plt.show()
```

```python
# include more data in our plot
# show predictions ONLY, not data
sns.relplot(
    x='avg_temp',
    y=best_model.predict(citibike[['avg_temp', 'precip', 'weekend', 'any_snow']]),
    data=citibike,
    size='precip',
    hue='weekend',
    col='any_snow',
    height=8,
    aspect=1
)
plt.show()
```
