# 3. Preregistration Statement

## 3.1. Single-variable Linear Regression

We will first **carry out single-variable regressions for each variable** that might influence the percentage of women inventors across different countries within the same time period (2010-2014). 

Predicting variables:
- Gross enrolment ratio, tertiary, gender parity index (GPI)
- Percentage of graduates from Science and Technology programmes in tertiary education who are female
- Percentage of women declaring that they have access to training on how to start a business
- Percentage of women declaring that they have access to money to start a business
- Percentage of women preferring taking a risk and starting a new business to working for someone else
    
Therefore, there will be a total of five linear regressions. We will calculate the **correlation coefficient** and **R square (coefficient of determination)** for each linear regression to examine whether there is a strong linear correlation between these variables and our dependent variable, which is the percentage of women inventors.

We will also calculate the **p-values** for each regression to check their significance, together with a **hypothesis test**.

With the results from single-variable regressions, we can then figure out which variables to include in our next step, which is the multi-variable linear regression. To ensure a reliable multi-variable linear regression, we will **exclude the variables that have the highest correlation and r square value**.

## 3.2. Multivariable Linear Regression

In reality, many powerful linear regression models are not limited to single input variables, given that we are living in a complex world with countless variables. **We will then carry out multivariable regressions** for those selected variables that might influence the percentage of women inventors across different countries within the same time period (2010-2014). For those predicting variables that are highly correlated and might have a causal effect on each other, **we will exclude the ones with the highest correlation and r square value**. For example, “Percentage of women declaring that they have access to training on how to start a business” and “Percentage of women declaring that they have access to money to start a business” might be highly correlated, and adding both variables into the multivariable regression might not bring about much new information compared to introducing only one of them.

After we obtained a multivariable regression model, we will calculate its **correlation coefficient** and **r square score** to examine whether our multivariable regression has a strong explanatory power (i.e., explaining the percentage of women inventors in countries that are included in our model).

We will also calculate the **p-value** for the multivariable regression to check its significance, together with a **hypothesis test**.

We can **compare the r squares for the single regressions and multivariable regression** to see which has a higher coefficient of determination, and therefore we are able to predict the relationship between the variables more accurately.

### 4.2.1. Single-variable Linear Regression

Predicting variables:
- Gross enrolment ratio, tertiary, gender parity index (GPI)
- Percentage of graduates from Science and Technology programmes in tertiary education who are female
- Percentage of women declaring that they have access to training on how to start a business
- Percentage of women declaring that they have access to money to start a business
- Percentage of women preferring taking a risk and starting a new business to working for someone else

We will build 5 single-variable linear models to predict 'Share of Women Inventors (%)' from the variables above.

Then, we will calculate the **correlation coefficient** and **R square (coefficient of determination)** for each linear regression to examine whether the correlation is strong.

We will also calculate the **p-values** for each regression to check their significance, together with a **hypothesis test**.

With the results from single-variable regressions, we can then figure out which variables to include in our next step, which is the multi-variable linear regression. To ensure a reliable multi-variable linear regression, we will **exclude the variables that have the highest correlation and r square value**

```python
# Linear Regression: Female Inventors (%) vs GPI

gpi_dropna = final.dropna(subset=['Country', 'Gross enrolment ratio, tertiary, gender parity index (GPI)'])
grouped_gpi = gpi_dropna.groupby('Country')

X = grouped_gpi['Gross enrolment ratio, tertiary, gender parity index (GPI)'].mean()
Y = grouped_gpi['Share of Women Inventors (%)'].mean()

regressor = LinearRegression()
regressor.fit(X.values.reshape(-1,1), Y)

sns.regplot(x=X, y=Y)
plt.xlabel('GPI')
plt.ylabel('Share of Women Inventors (%)')
plt.title("Female Inventors (%) vs GPI [LinearRegression]", size=20)
plt.show()

print('Predictor: Gross enrolment ratio, tertiary, gender parity index (GPI)')
print('Slope:', regressor.coef_)
print('Intercept:', np.round(regressor.intercept_, decimals=4))
print('Correlation Coefficient:', np.round(np.corrcoef(X,Y)[0,1], decimals=4))
print('R square (coefficient of determination):', np.round(regressor.score(X.values.reshape(-1,1), Y), decimals=4))
```

### 4.2.2. Multivariable Linear Regression

**We will then carry out multivariable regressions** for those selected variables that might influence the percentage of women inventors across different countries within the same time period (2010-2014). For those predicting variables that are highly correlated and might have a causal effect on each other, **we will exclude one of them**.

After we obtained a multivariable regression model, we will calculate its **correlation coefficient** and **r square score** to examine whether our multivariable regression has a strong explanatory power (i.e., explaining the percentage of women inventors in countries that are included in our model).

We will also calculate the **p-value** for the multivariable regression model to check its significance, together with a **hypothesis test**.

We can **compare the r squares for the single regressions and multivariable regression** to see which has a higher coefficient of determination, and therefore we are able to predict the relationship between the variables more accurately.

### 5.1. Evaluation for Gender Parity Index (GPI) and the Share of Women inventors

```python
# Linear Regression: Female Inventors (%) vs GPI

gpi_dropna = final.dropna(subset=['Country', 'Gross enrolment ratio, tertiary, gender parity index (GPI)'])
grouped_gpi = gpi_dropna.groupby('Country')

X = grouped_gpi['Gross enrolment ratio, tertiary, gender parity index (GPI)'].mean()
Y = grouped_gpi['Share of Women Inventors (%)'].mean()

regressor = LinearRegression()
regressor.fit(X.values.reshape(-1,1), Y)

sns.regplot(x=X, y=Y)
plt.xlabel('GPI')
plt.ylabel('Share of Women Inventors (%)')
plt.title("Female Inventors (%) vs GPI [LinearRegression]", size=20)
plt.show()

print('Predictor: Gross enrolment ratio, tertiary, gender parity index (GPI)')
print('Slope:', regressor.coef_)
print('Intercept:', np.round(regressor.intercept_, decimals=4))
print('Correlation Coefficient:', np.round(np.corrcoef(X,Y)[0,1], decimals=4))
print('R square (coefficient of determination):', np.round(regressor.score(X.values.reshape(-1,1), Y), decimals=4))
```

Looking at the slope and r square values we obtained, we can say that the single-variable regression model for evaluating the relationship between "Gross enrollment ratio, tertiary, gender parity index (women to men)" and the "Share of Women inventors" is not of a good fit. Therefore, this regression model is not suitable for prediction.

```python
import statsmodels.formula.api as smf

df = pd.DataFrame({"x": X, "y": Y})
model = LinearRegression().fit(df[["x"]], df["y"])
sm_model = smf.ols("y ~ x", data=df)
result = sm_model.fit()
result.summary()
```

From the chart above, we can see that the **p-value is 0.69**, and the confidence level is [-9.432, 6.328]. Since p-value is significantly larger than 0.05, we **cannot reject the null hypothesis**. Therefore, we can conclude that there is probably no relationship between Gender Parity Index (GPI) and the Share of Women inventors.

Though 48% is not a very high precentage, it has stronger prediction than other single and multivariable regressions with values such as 37.6%, 22.9%, and 27.5%. 

It is hard to say whether the multivariable regression model excluding 'Access to Training' is able to predict our desirable dependent variable well, but this is the best and most convincing result we can get from the analysis above.

# 6. Interpretation and Conclusions

Over the course of our data analysis, we first carried out a preliminary exploration of the dataset using histograms, scatterplots, and lineplots, to have a glimpse of what our data looks like overall. We also calculated the mean and standard deviation for each of the variables we selected among the OECD countries. From this preliminary exploration, we found out for the OECD countries where female respondents stated not having access to money, the percentages of female inventors are likely to be higher; for the OECD countries where the female survey respondents state preference of taking risks and not working for someone else, the shares of women inventors are also likely to be higher. 

Keeping this trend in mind, we further conducted deeper analysis with linear regression models, and finally used hypothesis tests to evaluate the result we obtained. 

We carried out **single-variable linear regression models** first to examine the relationship between each independent variable and the dependent variable, share of women inventors. We came to the below conclusion:

1. There is a negative weak correlation between "Gross enrolment ratio, tertiary, gender parity index" i.e. GPI and the Share of Women inventors, since we have a **slope of -1.55**. As the **coefficient of determination is 0.0054, this indicates that only 0.54% of the data fall within the regression line.**


2. There is a positive weak correlation between Percentage of graduates from Science and Technology programmes in tertiary education who are female (%) and the Share of Women inventors, since we have a **slope of 0.134**. As the **coefficient of determination is 0.0322, this indicates that only 3.22% of the data fall within the regression line.** 


3. There is a negative weak correlation between Percentage of graduates from Science and Technology programmes in tertiary education who are female (%) and the Share of Women inventors, since we have a **slope of -0.0682**. As the **coefficient of determination is 0.0870, this indicates that only 8.70% of the data fall within the regression line.**


4. There is a negative moderate correlation between Percentage of women declaring that they have access to money to start a business (%) and the Share of Women inventors, since we have a **slope of -0.176**. As the **coefficient of determination is 0.2291, this indicates that only 22.91% of the data fall within the regression line.** 


5. There is a positive moderate correlation between Percentage of women preferring taking a risk and starting a new business to working for someone else (%) and the Share of Women inventors, since we have a **slope of 0.165**. As the **coefficient of determination is 0.2756, this indicates that only 27.56% of the data fall within the regression line.**

We found out that only the last two results are of moderate significance, after carrying out hypothesis tests for each of them. We got **p-values of 0.01 and 0.004 for the fourth and fifth conclusion above correspondingly.** Since they are both smaller than 0.05, we can say that there is a relationship between the two variables and the share of women inventors. However, the other three failed to reject the null hypothesis, given that their p-values are 0.69, 0.334, and 0.127 correspondingly, which are all way larger than 0.05.

We then processed our data by dropping all the NaN's to conduct a **multivariable linear regression model**, in order to get the most reliable prediction method for the dataset. For the multivariable regression, we decided to exclude the variable "Percentage of women declaring that they have access to training on how to start a business (%)" since it has a strong correlation with another variable, and **the r square value for this particular multivariable model is the highest, with a value of about 0.48. This indicates that we can predict about 48% of the variation in the share of women inventors with the selected variables.**

However, we are not so confident about the conclusions we drew from the data analysis, because the hypothesis tests suggest that our model does not have an extremely strong ability in predicting the change in our dependent variable, with a relatively low p-value. 

Still, we were able to figure out an overarching view of the relationship between the independent variables and the dependent variable we are interested in. To apply the results obtained from our more reliable models to a wider context of real-life application, we can conclude that "Gross enrolment ratio, tertiary, gender parity index" i.e. GPI, "Female Sci-Tech Graduates," and "Access to Training" are probably the three variables with the smallest absolute value of correlation coefficients with "Share of Women Inventors" among OECD countries, as both single and multivariable linear regression models show small correlation between these variable and the "share of women inventors."

From the multivariable regression that excludes "Access to Training", we can also see that Gender Parity Index (GPI) might be the strongest influencing factor of the change in "Share of Women Inventor", with a coefficient of 0.46. **However, this may also be the consequence of a property of GPI**: Gender Parity Index is a relatively small index ranging between 0 and 2, while all the other variables are percentages ranging from 0 (%) to 100 (%). For the other variables, "Female Sci-Tech Graduates", "Access to Money" and "Positive Attitude to Risks", the coefficients are 0.13, -0.15, and 0.15 accordingly, indicating moderate influence on the percentage of female inventors.

From the data analysis results of correlation above, we conclude that "Gross enrollment ratio of tertiary education (women to men)" i.e. GPI and "Female Sci-Tech Graduates (%)" are **not** the most important factors that influence or contribute to the change in the share of women inventors. **Instead, "Access to Money" (which also suggests access to training) and "Positive Attitude towards Entrepreneurial Risks" take the crucial roles.**

We think this conclusion makes sense, as it infers the importance of subjective initiative among women in developed countries. In developed countries, Access to Money may not be the most critical factor in influencing women's decision in starting an invention, as there is a negative association with the share of women inventors. This is possibly because people from these groups have already lived a relatively medium-to-high-standard life, compared to citizens of developing countries, and therefore they have access to corresponding **money and training resources**. However, what actually influences the share of women inventors is whether they have the **will** to invent something through actual endeavor academically. **If this initiative is absent, then access to ample resources (whose alias might be "steady, cozy lifestyle") may become the obstacle to risky inventions.**

Our conclusion matters because it suggests that **to improve the share of women inventors, we should put more attention to encouraging women to take initiative in trying and learning related knowledge, and change their attitudes towards inventive challenges**. The historical data suggests that the fluctuation in share of women inventors was possibly influenced by GPI, as well as "Percentage of female graduates in sci-tech majors" and "Positive attitudes towards invention", we can follow the direction and predict future trend. Hopefully, **gender equality can be achieved or improved**--not only in OECD countries selected in this study, but also in all developing countries--if more efforts are invested into these areas, and **more excellent female inventors will appear on the stage**.

# 7. Limitations

One limit of our dataset is that we didn't include the public opinions on women inventors (or entrepreneurs), which can be an important externality that influence the share of women inventors. Though we thought about this critical factor, we didn't include it since public opinions are hard to be quantized. We didn't find suitable data source that include data from the countries and timeframe we are researching on either.

In terms of harm and gap, our dataset is also limited by the choice of countries. We only focus on OECD member countries, which are all already developed. Therefore, our dataset can only reflect the situation of women inventors in developed countries, but we are unable to explore that of devloping and undert-developed ones. The population of developed countries only make up of a small proportion of the world population, so our dataset cannot reflect the overall realistic situation of women. If we want to actually figure out underlying trend and reasons that influence share of owmen inventors, we should go beyond just looking at developed countries. We should in turn focus more on missing data and try to collect data from all over the world. However, due to the limited time and resource, we only focus on existing data on developed countries, hoping to get a preliminary understanding of the topic of women inventors. Another associating limitation due to choice of countries is that there are only 33 OECD member countries, making our dataset relatively small in the scale. 

Last but not least, our dataset is also limited in the timeframe we chose. We only focus on data from 2010-2014, which is only a 5-year period. This period of time is also over 5 year from now. Though we believe this is a representative and meaningful period of time, our dataset is relatively small in its scale. There may be some other trends and influencial factors that might influence share of women inventors, but we cannot include them.

In all, though we have included multiple valuable variables that may influence share of women inventors that can potentially bring interesting discussion and analysis of the issue, our dataset size is not massive enough to conclude a comprehensive result.
