In this section, we are conducting sentimental analysis by using what we have learned in class regardin logistic regression. We referred to the article on Kaggle and nltk article (CHANGE PHRASING) to assist us to better understand and use the package of sentimental analysis.

Since sentimental analysis is only a small part of our analysis, we decided to reference the sentimental analysis using logistic regression on Kaggle.
```python
nltk.download('punkt')
nltk.download('vader_lexicon')
analyzer = SIA()
```

```python
def sentiment_analyzer_score(sentence):
    score = analyzer.polarity_scores(sentence)
    if score['compound']>0.05:
        return str('positive')
    if score['compound']<-0.05:
        return str('negative')
    else:
        return str('neutral')
```

```python
def sentiment_append(text):
    moods=[]
    for mess in text:
        mood=sentiment_analyzer_score(mess)
        moods.append(mood)
    return moods
```

### Title Sentimental Signal Analysis

Similar to three technical stock indictors such as Parabolic SAR, if there is a change in the indicator, it would suggest a buy/sell signal. We thought it would be insightful to conduct an signal analysis using the sentimental score we have calculated.

Since the congregated daily sentiment is mainly positive, we believe it would be intuitive to calculate the percentage change of the daily sentiment. 

```python
sentiment_list = []
previous = 0 
preprevious = 0

for i in range(len(gme)):
    if i == 0:
        sentiment_list.append("buy")
    elif gme.title_sentiment[i] < previous and preprevious < previous: 
        sentiment_list[-1] = "sell"
        sentiment_list.append("")
    elif gme.title_sentiment[i] > previous and preprevious > previous: 
        sentiment_list[-1] = "buy"
        sentiment_list.append("")
    else: 
        sentiment_list.append("")
    preprevious = previous
    previous = gme.title_sentiment[i]
  
    
print (sentiment_list)
gme['sentiment_signal'] = sentiment_list
gme.head(10)
```

Now that we have drafted the list of transaction dates we want to compare it to the best case scenario dates. We will do this by checking how many times did the dates of recommended transactions match the best case scenarios dates. 

Based on the lists that we calculated above, we can see that the length of recommended transactions varies greatly depending on the recommendation system. We can already begin to predict that the filter rule likely has the greatest odds of having the highest accuracy in predicting the correct buy-sell dates based on the sheer number of recommended transactions. 

There are two reasons for why the indicator would not fully match the best case scenario. The first reason is that the indicator missed the date by a few days before or after. The second reason is that the indicator recommended fewer transactions than the best case did. While in the reality we may want to pay closer attention to the number of recommended back and forth trades in a single share price, for simplicity we will ignore transaction costs. This suggests that even before we delve into deeper analysis, the macd and parabolic indicators will likely be at a disadvantage to the filter rule. 

Here we have calculated the accuracy two ways. First, we checked the number of correct protections divided by the total number of best case scenario recommended trades. This will entirely ignore transaction costs. The second lens that we can analyze is by dividing the correct number of observations by the total number of recommended transactions by that particular indicator. This way we are not completely discounting indicators that offer lower transactions per period. 

```python
# find the number of dates that the buy transactions and sell transaction dates matched between the indicator and the best case
psar_correct = len(set(psar_buy)& set(bc_buy)) + len(set(psar_sell) & set(bc_sell)) 
macd_correct = len(set(macd_buy)& set(bc_buy)) + len(set(macd_sell) & set(bc_sell)) 
filter_correct = len(set(filter_buy)& set(bc_buy)) + len(set(filter_sell) & set(bc_sell)) 
sent_correct = len(set(sent_buy)& set(bc_buy)) + len(set(sent_sell) & set(bc_sell)) 

print("Parabolic SAR matching number of dates: {}".format(psar_correct))
print("MACD matching number of dates: {}".format(macd_correct))
print("Filter rule matching number of dates: {}".format(filter_correct))
print("Sentiment Analysis matching number of dates: {}".format(sent_correct))
```
